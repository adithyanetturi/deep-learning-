```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
## creating an ANN model using tensorflow and keras package
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
## loading the iris dataset which is available in sklearn library
```

```python
iris.feature_names
#using this we can get features
```

```python
iris.data
```

```python
iris.target_names
```

```python
iris.target
```

```python
from sklearn.model_selection import train_test_split
#spliting the data
x_train,x_test,y_train,y_test = train_test_split(iris.data,iris.target,test_size=0.20,random_state=1)
```

#### creating ANN model

```python
model = Sequential()
input_layer = Dense(4,input_shape = (4,))
model.add(input_layer)
hidden_layer = Dense(4,activation='relu')
model.add(hidden_layer)
output_layer = Dense(3,activation='softmax')
model.add(output_layer)
```

#### compile 

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer = 'adam',metrics = ['accuracy'])
# since it is a classification problem we using crossentropy
```

```python
model.fit(x_train,y_train, epochs=300,batch_size = 4)
#fiting the model
```

```python
val_loss,val_acc = model.evaluate(x_test,y_test)
print(val_loss,val_acc)
#printing the accuracy
```

```python
import numpy as np
arr = np.array([[5.1,3.5,1.4,0.2]])
result = model.predict(arr)
print(np.argmax(result))
#rtesting with sample data 
```
